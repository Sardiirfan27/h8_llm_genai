## PyMuPDF

Selanjutnya kita akan coba menggunakan PyMuPDF (fitz) yang dapat membaca tabel dan mengakses koordinat teks.

````python
import fitz  # PyMuPDF
from tabulate import tabulate

doc = fitz.open(pdf_path)
doc_output = ""

for page in doc:

    elements = []

    # 1️⃣ Detect tables
    tables = page.find_tables()
    table_bboxes = [table.bbox for table in tables.tables]

    # 2️⃣ Get text blocks
    blocks = page.get_text("blocks")

    for block in blocks:
        x0, y0, x1, y1, text, *_ = block

        # cek apakah block berada dalam area tabel
        inside_table = False
        for tb in table_bboxes:
            tx0, ty0, tx1, ty1 = tb
            if (x0 >= tx0 and x1 <= tx1 and
                y0 >= ty0 and y1 <= ty1):
                inside_table = True
                break

        if not inside_table:
            elements.append({
                "type": "text",
                "top": y0,
                "content": text.strip()
            })

    # 3️⃣ Add tables
    for table in tables.tables:
        elements.append({
            "type": "table",
            "top": table.bbox[1],
            "content": table.extract()
        })

    # 4️⃣ Sort by vertical position
    elements = sorted(elements, key=lambda x: x["top"])

    # 5️⃣ Render to markdown
    for el in elements:
        if el["type"] == "text":
            doc_output += el["content"] + "\n\n"
        else:
            table_data = el["content"]
            headers = table_data[0]
            rows = table_data[1:]
            md_table = tabulate(rows, headers=headers, tablefmt="github")
            doc_output += md_table + "\n\n"

print(doc_output)
````

Hasil di atas tampak lebih rapi dan terdapat tabel yang sesuai dengan tata letak yang dihasilkan dalam sumber PDF. Agar lebih rapi lagi, jika ingin menghapus line break berlebih, maka dapat menggunakan cara sebagai berikut

````python
import re
doc_output = re.sub(r'\n{3,}', '\n\n', doc_output) # mengganti 3 atau lebih newline dengan 2 newline
doc_output = doc_output.strip()

print('total karakter:', len(doc_output))
print(doc_output)
````

# Chunking Strategies

## Fixed Chunking

Pada bagian ini, kita akan mencoba melakukan fixed chungking dengan ukuran chunk sebesar 500 karakter. Kita akan mencoba membuat fungsi custom tanpa menggunakan langchain. Pada fungsi tersebut, kita akan melakukan chunking tanpa overlap.

`Overlap` berarti tiap chunk mundur sebanyak karakter yang ditentukan agar konteks tidak terpotong

````python
# chunking tanpa overlap dengan ukuran tetap tanpa overlap
def fixed_chunk(text, chunk_size=500):
    chunks = []
    for i in range(0, len(text), chunk_size):
        chunk = text[i:i+chunk_size]
        chunks.append(chunk)
    return chunks


# terapkan fungsi chunking pada teks markdown yang dihasilkan
fixed_chunks = fixed_chunk(md_text, 500)

print("Jumlah chunk:", len(fixed_chunks))
print("Contoh chunk pertama:\n")
print(fixed_chunks[0])
````

````python
print("Contoh chunk kedua:\n")
print(fixed_chunks[1])
````

Selanjutnya kita akan mencoba melakukan fixed chunking dengan overlap.

````python
def fixed_chunk_overlap(text, chunk_size=500, overlap=100):
    chunks = []
    step = chunk_size - overlap

    for i in range(0, len(text), step):
        chunk = text[i:i+chunk_size]
        chunks.append(chunk)

        if i + chunk_size >= len(text):
            break

    return chunks


overlap_chunks = fixed_chunk_overlap(md_text, 500, 100)

print("Jumlah chunk:", len(overlap_chunks))
print("Contoh chunk pertama:\n")
print(overlap_chunks[0])
````

````python
print("Contoh chunk kedua:\n")
print(overlap_chunks[1])
````

Selanjutnya kita akan coba dengan langchain untuk melakukan text splitter dan chunking.

````python
from langchain_text_splitters import CharacterTextSplitter

splitter = CharacterTextSplitter(
    chunk_size=500,      # setiap chunk maksimal 500 karakter
    chunk_overlap=100,    # overlap 100 karakter
    separator="\n\n" # ubah separtor menjadi separator="", jika ingin mengabaikan pemisah antar chunk.
)

chunks = splitter.split_text(md_text)

print("Total chunk:", len(chunks))
print("Contoh chunk pertama (total karakter):", len(chunks[0]),'\n')

print(chunks[0])
````

CharacterTextSplitter: adalah metode untuk memecah teks berdasarkan jumlah karakter dan kondisi separator. Kekurangan dari teknik ini adalah seringkali memotong di tengah kalimat atau paragraf jika jumlah karakternya habis , sehingga konteks semantik hilang.

Selain CharacterTextSplitter, kita dapat menggunakan RecursiveCharacterSplitter yang umumnya menghasilkan chunk yang lebih rapi dan mengatasi kekurangan dari CharacterTextSplitter, walaupun sebenarnya bisa juga menyesuaikan separatornya agar lebih rapi dan mirip seperti RecursiveChacaterTextSplitter.

````python
from langchain_text_splitters import RecursiveCharacterTextSplitter
rc_splitter = RecursiveCharacterTextSplitter(
    chunk_size=500,
    chunk_overlap=100,

)
chunks = rc_splitter.split_text(md_text)

print("Total chunk:", len(chunks))
print("Contoh chunk pertama (total karakter):", len(chunks[0]),'\n')
print(chunks[0])
````

## Recursive Chunking

Untuk melakukan Recursive Chunking kita dapat gunakan RecursiveCharacterTextSplitter dengan penambahan konfigurasi di separatornya.

Misalkan, kita telah mengetahui markdown kita ada heading dan kita mau menambahkan separator berdasarkan heading selain separator paragraf, newline, dan spasi. Heading utama (H1) dengan `"\n# "`, dan heading kedua (subheading) dengan `"\n## "`.


````python
from langchain_text_splitters import RecursiveCharacterTextSplitter

# Recursive splitter bisa memanfaatkan heading
splitter = RecursiveCharacterTextSplitter(
    chunk_size=500,
    chunk_overlap=100,
      separators=["\n## ", "\n# ", "\n\n", "\n", " ", ""]
)

chunks = splitter.split_text(md_text)

print("Total chunk:", len(chunks))
print("Contoh chunk pertama (total karakter):", len(chunks[0]),'\n')
print(chunks[0])
````

````python
for i, chunk in enumerate(chunks):
    print(f"Chunk {i} (total karakter):", len(chunk))
    print(chunk)
    print("-" * 50)
````

Jika kita ingin fokus hanya pada heading, efeknya adalah beberapa chunk akan memiliki ukuran karakter lebih dari 500 atau bisa diartikan bahwa seperator akan menjadi pertimbangan awal atau utama sebelumnya melihat chunk_size.

````python
from langchain_text_splitters import RecursiveCharacterTextSplitter

# Recursive splitter bisa memanfaatkan heading
splitter = RecursiveCharacterTextSplitter(
    chunk_size=500,
    chunk_overlap=100,
    separators=["\n## ", "\n# "] # fokus hanya pada heading untuk memisahkan chunk
)

chunks = splitter.split_text(md_text)

print("Total chunk:", len(chunks))
for i, chunk in enumerate(chunks):
    print(f"Chunk {i} (total karakter):", len(chunk))
    print(chunk)
    print("-" * 50)
````

Perhatikan pada chunk 1 total karakterknya adalah 872 karena syarat pemisahannya fokus ke H1 dan H2 tanpa melihat new line.

# Metadata & Hybrid Search

`Metadata` adalah suatu data atau informasi tambahan yang dapat membantu untuk menjelaskan, mengelompokkan, atau memberi konteks pada suatu dokumen.

Kegunaan metadata diantaranya yaitu:
- Membuat Search Lebih Presisi karena bisa filter dokumen atau data.
- Mengurangi noise pada hasil pencarian dengan mengeliminasi dokumen yang tidak relevan atau membatasi domain pencarian.
- Membantu dalam segmentasi data.


Selain metadata, kita dapat menggunakan hybrid search untuk melakukan pencarian dengan lebih akurat.
Hybrid search adalah gabungan dua metode pencarian berbeda secara scoring, biasanya:
- Dense retrieval (semantic search)
- Sparse retrieval (keyword search)

## ChromaDB

````python
import chromadb
from chromadb.config import Settings
from sentence_transformers import SentenceTransformer

# Inisialisasi model embedding dan ChromaDB client
model = SentenceTransformer("all-MiniLM-L6-v2")
client = chromadb.Client(
    Settings(
        persist_directory="./chroma_db",
        is_persistent=True
    )
)


# Buat collection baru di ChromaDB
collection = client.create_collection("ecommerce",
                                      metadata={"hnsw:space": "cosine"}
                                      )
````

Pada konfigurasi diatas, kita akan mencoba menggunakan cosine distance untuk menghitung jarak antar embedding.

````python
# Gabungkan nama + deskripsi jadi 1 document
documents = [
f"""Nama Produk: {row.nama_produk}.
Deskripsi: {row.deskripsi}
"""
for _, row in df.iterrows()
]

documents
````

````python
# Buat embedding
embeddings = model.encode(documents).tolist()
````

````python
# Insert data ke Chroma
collection.add(
    documents=documents,
    embeddings=embeddings,
    metadatas=metadatas,
    ids=df["id"].astype(str).tolist()
)
````

````python
# cek collection
print(collection.count())
display(collection.get())
````

Selanjutnya kita akan coba melakukan pencarian dengan filtering, misalkan User ingin laptop dengan harga <= 13 juta dan stok tersedia. Kita akan menggunakan `where` di ChromaDB dan menggunakan beberapa operator.

Operator yang sering muncul atau digunakan :

Operator Perbandingan:

| Operator | Artinya                    |
| -------- | -------------------------- |
| `$eq`    | equal (=)                  |
| `$ne`    | not equal (!=)             |
| `$gt`    | greater than (>)           |
| `$gte`   | greater than or equal (>=) |
| `$lt`    | less than (<)              |
| `$lte`   | less than or equal (<=)    |


Operator Logika:

| Operator | Artinya                          |
| -------- | -------------------------------- |
| `$and`   | Semua kondisi harus terpenuhi    |
| `$or`    | Salah satu kondisi terpenuhi     |


````python
query = "laptop ringan dan cocok untuk kerja"
query_embedding = model.encode([query]).tolist()


results = collection.query(
    query_embeddings=query_embedding,
    n_results=3,
    where={
        "$and": [
            {"kategori": "laptop"},
            {"harga": {"$lte": 13000000}},
            {"stok": True}
        ]
    }
)

for doc, meta, dist in zip(
    results["documents"][0],
    results["metadatas"][0],
    results["distances"][0]
):
    print("Document:")
    print(doc)
    print('metadata:')
    print("Brand:", meta["brand"])
    print("Harga:", meta["harga"])
    print("Rating:", meta["rating"])
    print("-" * 40)
````

### Hybrid Search di ChromaDB

Chroma versi cloud menyedikan API atau dukungan khusus untuk pencarian hybrid. Namun, disini kita tidak akan gunakan versi cloud. Kita akan coba mengkombinasikan keyword search (BM25) dengan semantic search (dense retrival dengan all-MiniLM-L6-v2).

Pertama-tama kita akan mencoba pencarian menggunakan semantic search.

````python
# Dense Search via Chroma dengan model All-MiniLM-L6-v2
query =  "Sepatu lari ringan untuk olahraga."
query_embedding = model.encode([query])

dense_results = collection.query(
    query_embeddings=query_embedding.tolist(),
    n_results=len(documents)
)

dense_docs = dense_results["documents"][0]
dense_distances = dense_results["distances"][0]
dense_docs
````

````python
dense_distances
````

Selanjutnya, kita akan mencoba mengurangi skor dari distance dengan 1. Tujuannya agar skor yang tinggi berada di atas skor yang rendah.

````python
import numpy as np
dense_scores = 1 - np.array(dense_distances)
dense_scores
````

Selanjutnya, kita akan mencoba pencarian menggunakan keyword search.

````python
from rank_bm25 import BM25Okapi

tokenized_docs = [doc.lower().split() for doc in documents]

print("Tokenized documents:")
for doc in tokenized_docs:
    print(doc)
````

````python
bm25 = BM25Okapi(tokenized_docs)
print("IDF:", bm25.idf)
print("Average doc length:", bm25.avgdl)
````

````python
bm25_scores = bm25.get_scores(query.lower().split())
````

Lakukan normalisasi skor agar berada dalam rentang 0-1

````python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

# normalisasi skor agar berada dalam rentang 0-1
dense_scores_norm = scaler.fit_transform(dense_scores.reshape(-1,1)).flatten()
bm25_scores_norm = scaler.fit_transform(bm25_scores.reshape(-1,1)).flatten()
````

Setelah melakukan normalisasi, kita akan mencoba mengkombinasikan skor dari semantic search dan keyword search kemudian melakukan rangking. Skor terbesar akan menjadi skor terbaik atau yang paling relevan dengan query yang diberikan. Untuk perhitungan ini kita akan menggunakan `Weighted Linear Combination` atau `Linear Score Fusion`.

````python
# Hybrid Fusion
alpha = 0.6

hybrid_scores = alpha * dense_scores_norm + (1 - alpha) * bm25_scores_norm

results = list(zip(documents, hybrid_scores))
results = sorted(results, key=lambda x: x[1], reverse=True)

print("=== HYBRID SEARCH ===\n")
for doc, score in results:
    print(f"{score:.4f} - {doc}")
````

Dari hasil diatas kita dapat melihat selisih skor yang cukup besar antara antara produk sepatu dan laptop sehingga kita dapat dengan mudah menentukan produk mana yang lebih relevan, bahkan skor dari sepatu lari ringan untuk olahraga adalah 1. Berbeda dengan sebelum kita melakukan hybrid search, produk laptop tetap cenderung memiliki nilai yang berdekatan dengan sepatu.

````python
from pinecone import Pinecone, ServerlessSpec
from pinecone_text.sparse import BM25Encoder
from sentence_transformers import SentenceTransformer
import numpy as np
````

Selanjutnya kita akan lakukan inisialisasi Pinecone dan membuat index baru.

````python
# Inisialisasi Pinecone client
PINECONE_API_KEY = "your-pinecone-api-key"
pc = Pinecone(api_key=PINECONE_API_KEY)
index_name = "hs-products"
````

````python
# hapus index pinecone
# pc.delete_index('latihan-semantic-search')
# pc.delete_index(index_name)
````

Di pinecone untuk melakukan hybrid search, karena menggunakan sparse value maka kita perlu mengatur metric menjadi `dotproduct`. Jika tidak diatur ke dotproduct maka akan error saat melakukan query atau search.

````python
# buat index baru jika belum ada
if index_name not in pc.list_indexes().names():
    pc.create_index(
        name=index_name,
        dimension=384,  # all-MiniLM-L6-v2 = 384
        metric="dotproduct",  # wajib dotproduct , karena gunakan sparse
        spec=ServerlessSpec(
            cloud="aws",
            region="us-east-1"
        )
    )

index = pc.Index(index_name)
````

Setelah berhasil membuat index baru, kita akan mencoba melakukan inisialisasi model dense (semantic) dan sparse (keyword).

````python
# Gabungkan nama + deskripsi jadi 1 document
documents = [
f"""Nama Produk: {row.nama_produk}.
Deskripsi: {row.deskripsi}
"""
for _, row in df.iterrows()
]
````

````python
#inisialisasi dense dan sparse
# dense embedding dengan SentenceTransformer
model = SentenceTransformer("all-MiniLM-L6-v2")

# sparse dengan BM25
bm25=BM25Encoder()
# Fit BM25 pada korpus teks
bm25.fit(documents)
````

Selanjutnya kita akan melakukan encoding untuk membuat dense vector dan sparse vector.

````python
# Generate Dense Vectors
dense_vectors = model.encode(documents, show_progress_bar=True).tolist()

# Generate Sparse Vectors
sparse_vectors = bm25.encode_documents(documents)
````

Selanjutnya kita akan melakukan upsert ke Pinecone. Karena pinecone tidak menyimpan deksripsi dan nama_produk secara langsung, maka kita perlu menyimpannya sebagai metadata agar mempermudah dalam diakses ketika melakukan query.

````python
data_to_upsert = []

for i, row in df.iterrows():
    data_to_upsert.append({
        "id": str(row["id"]),
        "values": dense_vectors[i],
        "sparse_values": sparse_vectors[i],
        "metadata": {
            "nama_produk": row["nama_produk"],
            "deskripsi": row["deskripsi"], # Penting untuk level production
            "kategori": row["kategori"],
            "harga": row["harga"],
            "brand": row["brand"],
            "stok": row["stok"],
        }
    })

# Upsert ke Pinecone
index.upsert(vectors=data_to_upsert)
````

Agar lebih rapi dan lebih akurat sesuai dengan kebutuhan, kita dapat membuat fungsi dan menerapkan Weighted Hybrid Search atau yang sering disebut sebagai Alpha Tuning yang berfungsi untuk mengontrol perbandingan antara hasil dari semantic search dan keyword search.

- Jika Alpha = 1.0: Anda melakukan pencarian 100% Semantic. Kata kunci spesifik seringkali terabaikan demi makna umum.
- Jika Alpha = 0.0: Anda melakukan pencarian 100% Keyword (BM25). Hanya dokumen dengan kata yang sama persis yang akan muncul.
- Jika Alpha = 0.5 (Ideal): Keduanya memiliki bobot yang sama. Ini adalah titik awal terbaik untuk General Purpose Search.

Referensi: [Hybrid Search - Pinecone](https://docs.pinecone.io/guides/search/hybrid-search)

````python
def hybrid_query(query, top_k=3, alpha=0.5, filter_dict=None):
    # Encode Query
    dense_q = model.encode(query).tolist()
    sparse_q = bm25.encode_queries(query)

    # Hybrid Scaling: (Dense * alpha) & (Sparse * (1 - alpha))
    dense = [v * alpha for v in dense_q]
    sparse = {
        'indices': sparse_q['indices'],
        'values': [v * (1 - alpha) for v in sparse_q['values']]
    }

    # Query ke Pinecone
    result = index.query(
        vector=dense,
        sparse_vector=sparse,
        top_k=top_k,
        include_metadata=True,
        filter=filter_dict
    )

    return result['matches']
````

````python
query='Sepatu lari ringan untuk olahraga.'
results = hybrid_query(query, top_k=5, alpha=0.5)

for match in results:
    print('score:', match["score"])
    print("nama produk:", match["metadata"]["nama_produk"])
    print("deskripsi:", match["metadata"]["deskripsi"])
    print('='*40)
````

Sekarang skornya berada pada rentang 0-1, skor yang rendah menandakan kemiripan yang rendah, sedangkan skor yang tinggi menandakan kemiripan yang tinggi.

Selanjutnya, kita dapat melakukan pencarian dengan menambahkan filter jika ingin lebih spesifik.
